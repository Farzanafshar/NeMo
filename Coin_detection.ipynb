```
import cv2 as cv
import matplotlib.pyplot as plt
```

```
img_orig = cv.imread("/content/drive/MyDrive/20230625_220310.jpg")
```

```
img = cv.cvtColor(img_orig, cv.COLOR_BGR2RGB)
```

```
imgplot = plt.imshow(img)
```

```
img.shape
```

```
img_crop = img_orig[2500:3000, 0:350]
```

```
imgplot = plt.imshow(img_crop)
img_gray = cv.cvtColor(img_crop, cv.COLOR_BGR2GRAY)
imgplot = plt.imshow(img_gray)
```

```
img_gray.shape
```

```
_, img_thresh = cv.threshold(img_gray, 128, 255, cv.THRESH_BINARY)
```

```
imgplot = plt.imshow(img_thresh)
```

```

contours, hierarchy = cv.findContours(img_thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
```

```
len(contours)
```

```
import numpy as np
temp = np.zeros((img_gray.shape[0], img_gray.shape[1]))
```

```
temp.shape
```

```
img_cont = cv.drawContours(image=temp, contours=contours, contourIdx=-1, color=(255, 255, 255), thickness=2, lineType=cv.LINE_AA)
```

```
imgplot = plt.imshow(img_cont)
```
